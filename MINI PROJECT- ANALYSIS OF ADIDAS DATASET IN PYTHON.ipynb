# EXPLORATORY ANALYSIS ON ADIDAS MARKET DATASET

1.IMPORTING THE LIBRARIES:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2.IMPORTING THE DATASET:

```python
data=pd.read_csv('Adidas Final.csv')
data
```

The dataset contains the product sold in different region, the total number of sales, its operating margin, sales method and units sold.

3.DROPPING COLUMNS THAT ARE NOT REQUIRED FOR ANALYSIS:

```python
data.drop(['Retailer'],axis=1,inplace=True)
data.drop(['Retailer ID'],axis=1,inplace=True)
data.drop(['Invoice Date'],axis=1,inplace=True)
data.drop(['Operating Margin'],axis=1,inplace=True)
data
```

Now, let us find the different categories present under each columns to explore the dataset.

4.DIFFERENT TYPE OF PRODUCT CATEGORIES:

```python
categories = data['Product'].unique()
print(categories)
```

Therefore, the Adidas sold 3 different products each for men and women namely Street Footwear, Atheltic Footwear and Apparel.

5.VISUALIZATION OF THE PRODUCT CATEGORY:

```python
data['Product'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

This pie chart shows that there is an equal number of sales in Women's Street Footwear and Women's Atheletic Footwear and similar equal sale is observed in Men's Atheltic  Footwear and Men's Street Footwear. On comparison, Men's Street and Athletic footwear have been sold higher than Women's Street and Athletic footwear, while Women's Apparel is sold more than Men's Apparel.

6.IDENTIFYING THE DIFFERENT REGION, CITY AND STATE CATEGORIES:

```python
categories_1 = data['Region'].unique()
categories_1
```

7.USING BAR PLOT:

```python
sns.countplot(x ='Region', data = data)
```

It is clear from the graph that West Region has the highest number of sales followed by Northeast. Southeast has the least number of sales.

```python
categories_3= data['City'].unique()
categories_3
```

8.USING HORIZONTAL BAR CHARTS:

```python
sns.countplot(y ='City', data = data, orient='h')
```

9.DIFFERENT SALES METHOD:

```python
categories_4= data['Sales Method'].unique()
categories_4
```

There are 4 types of Sales Method- In-Store, Outlet and online.

```python
sns.countplot(x ='Sales Method', data = data)
```

The graph depicts that more number of sales has been through online and least through In-Store sales.

10.CREATION OF FREQUENCY TABLE:

Let us construct frequency table for Products, to find the total number of counts in each category.

```python
frequency_table = data['Product'].value_counts()
print(frequency_table)
```

11.PLOTTING FREQUENCY TABLE:

Let us plot the frequency table 

```python
plt.figure(figsize=(8, 4))
frequency_table.plot(kind='bar')
for i, count in enumerate(frequency_table):
    plt.text(i, count + 1, str(count), ha='center', va='bottom')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.title('Frequency Table')
plt.xticks(rotation=45)  
plt.show()
```

12.CREATION OF TWO WAY TABLE:

The two way table gives the total number of each type of footwear sold in each region respectively.

```python
two_way_table = pd.crosstab(data['Product'], data['Region'])
print(two_way_table)
```

As we know West Region has the highest number of sales, Men's Atheltic Footwear was sold the highest in overall.

13.PLOTTING TWO WAY TABLE:

```python
plt.imshow(two_way_table, cmap='YlGnBu', interpolation='nearest')
plt.colorbar()
plt.xlabel('Region')
plt.ylabel('Product')
plt.title('Two-Way Table')
plt.show()
```

14.PLOTTING HEATMAP:

```python
sns.heatmap(pd.crosstab(data['Product'], data['Region']))
```

15.PLOTTING CLUSTER MAP:

```python
sns.clustermap(pd.crosstab(data['Product'], data['Region']))
plt.show()
```

Thus, the Adidas Sales Dataset is analyzed using visualization techniques. We can infer that Men's Apparel was the top sold product among others and West Region had the top sales in all its product categories compared to other region. Also,comparison among other products and regions were also done using tables and charts.
